# Python Dictionaries

## looping through dictionariess

```python
dist1 = {
    'name' : 'Ramu',
    'age' : 23,
    'sex' : True,
    'study' : 'B.Sc.CSIT'
}

for x in dist1:
    print("keys",x,end=" ")

print("")

for x in dist1.values():
    print("vaues",x,end=" ")
```

```python
fruits = ["apple","ball","apple","ball","cat","mango"]

## get unique items using set
# mylist = list((set(fruits)))

my_dict = {
    'apple':0,
    'ball':0,
    'cat':0,
    'mango':0,
}
# print(mylist)

for x in fruits :
    if x == my_dict['apple']:
        my_dict['apple']+=1
    elif x == my_dict['ball']:
        my_dict['ball']+=1
    elif x == my_dict['cat']:
        my_dict['cat']+=1

print(my_dict['apple'])
```
